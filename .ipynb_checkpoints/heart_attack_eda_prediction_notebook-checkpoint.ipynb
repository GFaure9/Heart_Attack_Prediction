# _Heart Attack Prediction_ :  EDA and Prediction

## 1. Dataset Description

The dataset was taken from https://www.kaggle.com/datasets/rashikrahmanpritom/heart-attack-analysis-prediction-dataset.
It is a dataset for heart attack classification.

The variables are:

<u> Continuous Variables </u>
* __age__ : age of the patient
* __trtbps__ : resting blood pressure (in mm Hg)
* __chol__ : cholestoral in mg/dl fetched via BMI sensor
* __talachh__ : maximum heart rate achieved
* __oldpeak__ : previous peak

<u> Categorical Variables </u>
* __sex__ : sex of the patient
* __exng__: exercise induced angina (1 = yes; 0 = no)
* __caa__: number of major vessels (0-3)
* __cp__ : chest pain type
  - 0= typical angina
  - 1= atypical angina
  - 2= non-anginal pain
  - 3= asymptomatic
* __fbs__ : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
* __rest_ecg__ : resting electrocardiographic results
  - 0= normal
  - 1= having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
  - 2= showing probable or definite left ventricular hypertrophy by Estes' criteria
* __slp__ : slope (0-2)
* __thall__ : thalium stress test result (0-3)
  
<u> Target Variable </u>
* __output__ : 0= less chance of heart attack 1= more chance of heart attack

### iv. Variables Selection and Correlations Analysis

We will start by plotting the correlation map between the continuous variables.

```python
cont_raw_data = raw_data[cont_var]

corr_matrix = cont_raw_data.corr()

# To show only the lower triangular matrix
mask = ~np.tril(np.ones_like(corr_matrix, dtype=bool))  # use triu for the upper part
masked_corr = corr_matrix.mask(mask)

# sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
sns.heatmap(masked_corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Continuous Variables')
plt.show()
```

Since Pearson's correlation coefficient is not really adapted for categorical variables, we will look at the frequency distribution of the combinations of pairs of categorical variables. We can also:

&#8209; perform a $\chi^2$ test of independence to assess if there is a significant association between two categorical variables;

&#8209; use association measures like Cramér's V or Theil's U to calculate the strength of association between categorical variables.


```python
# Categorical Variables
cat_var = ['sex', 'exng', 'caa', 'cp', 'restecg', 'fbs', 'slp', 'thall']
```

```python
# Frequency Distribution of Categorical Variables pairs

fig, axs = plt.subplots(len(cat_var) - 1, len(cat_var) - 1, figsize=(12, 12))

pairs = []

for i, vari in enumerate(cat_var):
    cat_var_reduced = cat_var.copy()
    cat_var_reduced.remove(vari)
    
    for j, varj in enumerate(cat_var_reduced):
        pairij = (vari, varj)
        pairji = (varj, vari)
        
        if (pairij not in pairs) and (pairji not in pairs):
            ax = axs[i][j]
            freq_tab = pd.crosstab(raw_data[vari], raw_data[varj])
            sns.heatmap(freq_tab, cmap='YlGnBu', annot=True, fmt='d', ax=ax)
#             ax.set_title('%s VS %s' % (vari, varj))
            
            pairs.append(pairij)
            pairs.append(pairji)

plt.suptitle('Frequency Distribution of Categorical Variables pairs')

# Adjust the layout and spacing
plt.tight_layout()

# Display the plots
plt.show()
```

```python
# Chi-Squared tests of Independence between pairs of Categorical Variables

cat_raw_data = raw_data[cat_var]

# Create an empty DataFrame to store the results
results = pd.DataFrame(index=cat_raw_data.columns, columns=cat_raw_data.columns)

# Loop over pairs of categorical variables
for var1 in cat_raw_data.columns:
    for var2 in cat_raw_data.columns:
        # Perform the chi-squared test
        contingency_table = pd.crosstab(cat_raw_data[var1], cat_raw_data[var2])
        chi2, p_value, _, _ = chi2_contingency(contingency_table)
        
        # Store the p-value in the results DataFrame
        results.loc[var1, var2] = p_value

# Convert p-values to numeric values
numeric_results = results.astype(float)

# To show only the lower triangular matrix
mask = ~np.tril(np.ones_like(numeric_results, dtype=bool))  # use triu for the upper part
masked_results = numeric_results.mask(mask)

# Create the heatmap
# sns.heatmap(numeric_results, annot=True, cmap='YlGnBu', fmt=".4f")
sns.heatmap(masked_results, annot=True, cmap='YlGnBu', fmt='.4f')
plt.title('Chi-Squared Test of Independence between Categorical Variables')
plt.show()
```

Here, the p-value indicates the probability of observing the data or more extreme data, assuming that the null hypothesis is true. The null hypothesis states that there is no association between the variables, meaning they are independent.

We remind how to interpret the $p-value$:

* $p-value < L$  (e.g., $L = 0.05$) $\Rightarrow$ we reject the null hypothesis. It suggests that there is sufficient evidence to conclude that there is a statistically significant association between the variables.

* $p-value \geq L$ $\Rightarrow$ we fail to reject the null hypothesis. It suggests that there is not enough evidence to conclude a significant association between the variables. However, it does not necessarily mean that there is no association; it just means that we do not have strong evidence to reject the null hypothesis.

We can make some observations from these correlation analyses:
* <u>Concerning continuous variables</u>:
    * The higher correlation coefficient is between the age and the maximum heart rate achieved ('talachh'); it is a negative correlation (-0.4), meaning that the older the person is, the lower will be its maximum heart rate. This seems to be physiologically consistent.    
    * The second higher correlation coefficient is also negative (-0.34) and is between the 'talachh' parameter and the 'oldpeak' parameter. This means that we expect the heart rate to be lower when the previous peak is higher.
    
* <u>Concerning categorical variables</u>:
    * We detect statistically significant association between the following variables:
        * ('sex', 'tall')
        * ('exng', 'cp')
        * ('exng', 'slp')
        * ('exng', 'thall')
        * ('caa', 'cp')
        * ('cp', 'slp')
        * ('cp', 'thall')
        * ('restecg', 'slp')
        * ('slp', 'thall')

### v. Dataset Pre-processing

#### Dealing with continuous variables: Standardization or Normalization ?

* <u>Normalization</u> consists in applying the following function to each variable $y$ (the resulting variable is between 0 and 1):

<center> $y_{normalized} = \frac{y - y_{min}}{y_{max} - y_{min}}$ </center>

$\rightarrow$ When ?

&#8209; If the distribution of the data is not Gaussian;

&#8209; For algorithms that do not assume any distribution of the data/ that are affected by the range of the features (ex: KNN, Neural Networks, K-means, SVM).


* <u>Standardization</u> consists in applying the following function to each variable $y$ (the resulting variable is centered around the mean with a unit standard deviation):

<center> $y_{standardized} = \frac{y - \sigma}{\mu}$ </center>


$\rightarrow$ When ?

&#8209; If the distribution follows a Gaussian distribution (but also when not);

&#8209; If we do not want the outliers to be affected;

&#8209; For tree-based algorithm, since they are not really sensitive to the scale of the features.

In our case, we will use another option which is the `RobustScaler` class of the `sickit-learn` package. This scaler allows to removes the median and scale the data in the range between 1st quartile and 3rd quartile (also known as the interquartile range): it is robust to outliers.

```python
# Create an instance of the RobustScaler
scaler = RobustScaler()

# Create the new dataset
prepro_data = raw_data.copy()

prepro_data[cont_var] = scaler.fit_transform(prepro_data[cont_var])

print('Before scaling continuous variables: \n \n', raw_data[cont_var].head())

print('\n \nAfter scaling continuous variables with RobustScaler: \n \n', prepro_data[cont_var].head())
```

#### Dealing with categorical variables: a quick overview

First, we must remind that there are two types of categorical variables: _nominal_ and _ordinal_ variables:

&#8209; Nominal = categories are labels or names without any numerical or meaningful order (__ex:__ 2 genders: male and female).

&#8209; Ordinal = categories with an inherent order or ranking (__ex:__ 3 levels: low, medium and high).

<br>

$\rightarrow$ What techniques can be used to pre-process these categorical variables ?

_For nominal variables mainly..._

* <u>One-hot encoding</u> consists in creating, for each variable, as many binary columns as categories. Hence, for each individual, we will have an observation that is 0 or 1 in each category. For instance, here, and if one individual had the value 2 (showing probable or definite left ventricular hypertrophy by Estes' criteria) for the variable 'rest_ecg', we will create 3 columns (0, 1 and 2) and the observations corresponding to the individual would be: 0, 0, 1.

    _Note:_ this method can increase considerably the number of features, hence exposing us the course of dimensionality. This is why, when we will have initially many features, we will generally combine one-hot encoding with PCA to reduce the cardinal of features space.


* <u>Dummy encoding</u> is similar to one-hot encoding, but one column is dropped to avoid introducing perfect multicollinearity among the encoded variables.

_For ordinal variables..._

* <u>Label encoding</u> consists in assigning a number to each category, following the order. For instance, for an hypothetical variable 'pain' with 3 levels of intensity (low, medium and high) we will have 3 numbers: 0, 1 and 2.

    _Note:_ it is better not to use it for nominal variables because we do not want an order to be introduced.


In our case, we will use dummy encoding since we have nominal variables.

```python
# Perform dummy encoding on categorical variables and update the new dataset

prepro_data = pd.get_dummies(prepro_data, columns=cat_var, drop_first=True)

print('Before encoding categorical variables: \n \n', raw_data[cat_var].head())

print('\n \nAfter encoding categorical variable with get_dummies: \n \n', prepro_data.head())
```

## 3. Prediction: ML Model

### i. Model Selection

We will test different models:
* Logistic Regression
* Decision Tree
* Random Forest
* XGBoost
* SVC (Support Vector Classifier)
* MLP (Multilayer Perceptron)
* CNN (Convolutional Neural Network)

We will start by defining a Python class to make it easier to tune and select our models.

```python
class ModelSelection:
    def __init__(self,
                 models: List[BaseEstimator],
                 xtrain: pd.DataFrame, ytrain: pd.DataFrame,
                 xtest: pd.DataFrame, ytest: pd.DataFrame,
                 params: List[dict]):
        self.models = models
        self.xtrain = xtrain
        self.ytrain = ytrain
        self.xtest = xtest
        self.ytest = ytest
        self.params = params
    
    def _kfold_best_params(self, model: BaseEstimator, param: dict, n_splits: int):
        kfold = KFold(n_splits=n_splits, shuffle=True, random_state=42)
        grid_search = GridSearchCV(model, param, cv=kfold)
        grid_search.fit(self.xtrain, self.ytrain)
        print(grid_search.best_score_)
        return grid_search.best_params_, grid_search.best_score_
    
    def rank_models(self, n_splits: list):
        scores = []
        names = ['accuracy', 'precision', 'sensitivity', 'F1', 'specificity']  #  'AUC-ROC']
        best_hyperparams = []
        for mdl, tested_prm in zip(self.models, self.params):
            prm = self._kfold_best_params(mdl, tested_prm, n_splits)[0]
            best_hyperparams.append(prm)
            mdl.set_params(**prm)
            
            mdl.fit(self.xtrain, self.ytrain)
            y_pred = mdl.predict(self.xtest)
#             y_scores = mdl.predict_proba(self.xtest)
            
            tn, fp, fn, tp = confusion_matrix(self.ytest, y_pred).ravel()
            specificity = tn / (tn + fp)
            scores.append([
                accuracy_score(self.ytest, y_pred),
                precision_score(self.ytest, y_pred),
                recall_score(self.ytest, y_pred),
                f1_score(self.ytest, y_pred),
                specificity,
#                 roc_auc_score(self.ytest, y_scores)
            ])
        
        return pd.DataFrame(data=np.array(scores), columns=names), best_hyperparams 
```

```python
# Model selection among : DecisionTree, RandomForest, GradientBoosting, SVC, MLP

models = [
#     LogisticRegression(),
    DecisionTreeClassifier(random_state=42),
    RandomForestClassifier(random_state=42),
    GradientBoostingClassifier(random_state=42),
    SVC(random_state=42),
#     MLPClassifier(random_state=42, max_iter=400),
    xgb.XGBClassifier(random_state=42),
    
]
params = [
#     {'C': [0.01], 'penalty': ['l1', 'l2']},
    {'max_depth': [3, 5, 7, 10, None], 'min_samples_split': [2, 5, 10, 20, 50]},
    {
        'n_estimators': [100, 200, 300],
        'max_depth': [1, 3, 5, 7, None],
        'min_samples_split': [2, 5, 10],
        'max_features': ['sqrt', 'log2']
    },
    {
        'learning_rate': [0.1, 0.01, 0.001],
        'n_estimators': [100, 200, 300],
        'max_depth': [1, 3, 5, 7, None],
        'min_samples_split': [2, 5, 10],
        'max_features': ['sqrt', 'log2']
    },
    {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': [0.1, 0.01, 0.001]
    },
#     {
#         'hidden_layer_sizes': [(50,), (100,), (200,), (50, 50), (100, 100)],
#         'activation': ['relu', 'logistic'],
#         'alpha': [0.0001, 0.001, 0.01]
#     },
    {
        'max_depth': [3, 5, 7],
        'learning_rate': [0.1, 0.01, 0.001]
    },
]

mdl_select = ModelSelection(models=models,
                            xtrain=X_train, ytrain=y_train, 
                            xtest=X_test, ytest=y_test,
                            params=params)

scores, hyperparams = mdl_select.rank_models(n_splits = 5)
i_best = np.argmax(scores['accuracy'])
best_model = models[i_best]
best_params = hyperparams[i_best]

print('Models Scores: \n', scores)
```

```python
best_model.set_params(**best_params)

best_model.fit(X_train, y_train)

y_pred = best_model.predict(X_test)

print('Accuracy: {} %'.format(np.round(accuracy_score(y_test, y_pred) * 100, 2)))

print(hyperparams)
```
